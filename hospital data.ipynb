10.Create a function called â€œcompute_hospital_meanâ€ to compute the mean test result_value. The function should have two arguments: test_code and hospital_ID. Merge the admin and lab data as appropriate or use previously created datasets. Return the result for test_code = KPL and hospital_ID = St. Michaelâ€™s Hospital.


```python
full.head()
```

```python
full.hospital_ID=full.hospital_ID.str.replace(pat=r'[^A-Za-z0-9]', repl=r' ', regex=True)  # replace all special symbols to space
```

```python
full.hospital_ID.unique()
```

```python
def compute_hospital_mean(test_code,hospital_ID):
    return full[(full.test_code==test_code) & full.hospital_ID.str.contains(hospital_ID)].result_value.mean()
```

```python
compute_hospital_mean('KPL' , "St  Michael s Hospital")
```
